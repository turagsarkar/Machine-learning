```python
#dataset
#import pandas

import pandas as pd
data=pd.read_csv('/content/Data_for_UCI_named.csv')
data
```

```python
#labelEncoding
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
data["stabf"]=label_encoder.fit_transform(data["stabf"])

data.head()
```

```python
X_corr=data.drop(['stab','stabf'],axis='columns')
y_corr=data['stabf']
```

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.ensemble import ExtraTreesClassifier
clf = ExtraTreesClassifier(n_estimators=100, random_state=0)
scores =cross_val_score(clf,X_corr, y_corr, cv=10)
scores
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=5)
scores =cross_val_score(clf,X_corr, y_corr, cv=10)
scores
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
from sklearn import svm
from sklearn.model_selection import cross_val_score
clf = svm.SVC(kernel='linear', C=1, random_state=42)
scores = cross_val_score(clf,X_corr, y_corr , cv=10)
scores
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import cross_val_score
sgd_clf = SGDClassifier(random_state=42)
scores = cross_val_score(sgd_clf,X_corr, y_corr, cv=10)
scores
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
from sklearn.ensemble import HistGradientBoostingClassifier
scores =cross_val_score(HistGradientBoostingClassifier(),X_corr, y_corr, cv=10)
scores
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
scores =cross_val_score(GradientBoostingClassifier(criterion='squared_error',n_estimators=10, learning_rate=0.05,max_depth=5, random_state=0),X_corr, y_corr, cv=10)
scores
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
from sklearn.linear_model import LogisticRegression
scores=cross_val_score(LogisticRegression(solver='liblinear',multi_class='ovr'),X_corr, y_corr, cv=10)
scores
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
from sklearn.ensemble import RandomForestClassifier
scores=cross_val_score(RandomForestClassifier(n_estimators=40),X_corr, y_corr, cv=10)
scores
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
from sklearn.ensemble import AdaBoostClassifier
scores=cross_val_score(AdaBoostClassifier(learning_rate = 0.05,n_estimators =10,random_state=42),X_corr, y_corr, cv=10)
scores
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=0)
scores=cross_val_score(clf,X_corr,y_corr, cv=10)
scores
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```
