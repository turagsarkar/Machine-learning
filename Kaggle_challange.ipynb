```python
import pandas as pd
import numpy as np
```

```python
data_read= pd.read_csv('flood.csv')
data_read.head()
```

```python
data_read.info()
```

```python
data_read.describe()
```

```python
data_read['FloodProbability'].unique()
```

```python
X= data_read.drop('FloodProbability',axis=1)
y= data_read['FloodProbability']
```

```python
X.info()
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
import pandas as pd

# Convert the NumPy array to a Pandas DataFrame
X_scaled_df = pd.DataFrame(X_scaled)

# Display the first few rows of the DataFrame
X_scaled_df.head()
```

```python
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.model_selection import cross_val_score
clf = ExtraTreesRegressor(n_estimators=100, random_state=0)
scores =cross_val_score(clf,X_scaled_df, y, cv=10)
scores
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
clf.fit(X_scaled_df,y)
clf.score(X,y)
```

```python
test_data= pd.read_csv('test.csv')
test_data.head()
```

```python
test_data_df=test_data.drop('id',axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
test_data_df = scaler.fit_transform(test_data_df)
```

```python
test_pred=clf.predict(test_data_df)
```

```python
submission=test_data[['id']]
submission['FloodProbability']=test_pred
submission.head()
```

```python
submission.to_csv('submission.csv',index=None)
```
